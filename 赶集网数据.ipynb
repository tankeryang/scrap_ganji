## 初始化mongodb

```python
client = pymongo.MongoClient('localhost', 27017)
ganji  = client['ganji']
url_list  = ganji['url_list']
url_link  = ganji['url_link']
item_info = ganji['item_info']
```

## 输出前300条数据的area字段值

```python
for i in item_info.find().limit(300):
    print(i['area'])
```

```python
#coding = utf8
area_list = []
for i in item_info.find():
    if i['area'][0] == '广州':
        area_list.append(i['area'][1]) 
area_index = list(set(area_list))
print(area_index)
```

```python
post_times = []
for index in area_index:
    post_times.append(area_list.count(index))
print(area_index, post_times, sep='\n')
```

```python
print(dict(zip(area_index, post_times)))
```

```python
def data_gen(types):
    length = 0
    if length <= len(area_index):
        for area, times in zip(area_index, post_times):
            data = {
                'name' : area,
                'data' : [times],
                'type' : types
            }
            yield data
            length += 1
```

```python
data_gen('column')
```

```python
for i in data_gen('column'):
    print(i)
```

```python
#coding = utf8
options = {
    'chart' : {'type' : 'column'},
    'title' : {'text' : 'Guangzhou each areas post counts compare'},
    'xAxis' : {'type' : 'category'},
    'yAxis' : {'title' : {'text' : 'post counts'}}
}

series = [i for i in data_gen('column')]
charts.plot(series, show='inline', options=options)
```
